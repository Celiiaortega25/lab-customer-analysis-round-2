```python
#As we don't have any value that is higher than 90% I'm not going to drop any column.
#And I'm going to replace all then NaN values with the mean
```

```python
lab2 = lab2.apply(lambda x: x.fillna(x.mean()) if x.dtype in [float, int] else x.fillna(x.mode().iloc[0]))
print(lab2)
```
